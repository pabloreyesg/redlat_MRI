```python
from nilearn.image import resample_img
import pylab as plt
import nibabel as nb
import numpy as np
```

```python
imagen_original = nib.load('t105.nii.gz')

downsampled_nii = resample_img(imagen_original, target_affine=np.eye(3)*1., interpolation='nearest')
#downsampled_niitri = resample_img(imagen_original, target_affine=np.eye(3)*1., interpolation='linear') no usar, solo de ejemplo

nib.save(downsampled_nii, 't1_1mm.nii.gz')
#nib.save(downsampled_niitri, 't1_1mm_lineal.nii.gz') solo para fines explicativos, no usar
```
